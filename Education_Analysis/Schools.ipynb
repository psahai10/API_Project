```python
# Import Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import requests
import json
import gmaps

# Census & gmaps API Keys
from config import gkey

# Configure gmaps
gmaps.configure(api_key=gkey)
```

```python
# Store filepath in a variable
Absent_rate = "Absent_Suspension_Rate.csv"
```

```python
#Read CSV file
Absent_rate_df = pd.read_csv(Absent_rate)
```

```python
#Show the header
Absent_rate_df.head()
```

```python
school_list = list(Absent_rate_df['Name'])
```

```python
#my_phrase = "Name"
base_url='https://maps.googleapis.com/maps/api/place/textsearch/json'
params = {
    "query": 'Name',
    "key": gkey
}

requests.get(base_url, params).json()
addresses = []
for school in school_list:
    query = school + 'in Los Angeles'
    params = {
    "query": query,
    "key": gkey
    }
    response = requests.get(base_url, params).json()
    try:
        addresses.append(response['results'][0]['formatted_address'])
    except:
        addresses.append('')
```

```python
addresses
```

```python
Absent_rate_df['school_address'] = addresses
Absent_rate_df
```
