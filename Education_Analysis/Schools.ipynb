```python
#my_phrase = "Name"
base_url='https://maps.googleapis.com/maps/api/place/textsearch/json'
params = {
    "query": 'Name',
    "key": gkey
}

requests.get(base_url, params).json()
addresses = []
for school in school_list:
    query = school + 'in Los Angeles'
    params = {
    "query": query,
    "key": gkey
    }
    response = requests.get(base_url, params).json()
    try:
        addresses.append(response['results'][0]['formatted_address'])
    except:
        addresses.append('')
```

```python
Absent_ratef = "Absent_Ratef.csv"
Absent_ratef_df = pd.read_csv(Absent_ratef)
Absent_ratef_df.head()
```

```python
url = 'https://maps.googleapis.com/maps/api/geocode/json'

add_list = Absent_ratef_df["school_address"]

geo_list = []
for address in add_list:
    

    params = {
             'address': address,
             'key': gkey
         }

    #Do the request and get the response data
    req = requests.get(url, params=params).json()

    #Use the first result
    result = req['results'][0]

    geodata = dict()
    geodata['lat'] = result['geometry']['location']['lat']
    geodata['lng'] = result['geometry']['location']['lng']
    geodata['address'] = result['formatted_address']
    geo_list.append(geodata)

    print('{address}. (lat, lng) = ({lat}, {lng})'.format(**geodata))
    
geo_list
```

```python
# Create it into Data Frame
Geo_df = pd.DataFrame(geo_list) 
Geo_df
```

```python
result = pd.concat([Geo_df, Absent_ratef_df], axis=1, sort=False)
result
```

```python
result['Suspension Rate'] = result['Suspension Rate'].str.replace('%', '')
result
```

```python
result['Absent Rate'] = result['Absent Rate'].str.replace('%', '')
```

```python
result['Absent Rate'] = pd.to_numeric(result['Absent Rate'], errors='coerce', downcast='float')
result
```

```python
# Store 'Lat' and 'Lng' into  locations 
locations = result[["lat","lng"]].astype(float)
absent_rate = result["Absent Rate"]
fig = gmaps.figure()

#fig.add_layer(gmaps.heatmap_layer(locations,weights=absent_rate))
# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=absent_rate,
                                  dissipating=False, max_intensity=.8,
                                  point_radius = .065)

heat_layer.dissipating = False
heat_layer.max_intensity = .8
heat_layer.point_radius = 0.065
heat_layer.opacity = 0.7
# Add layer
fig.add_layer(heat_layer)


#Display figure
fig
```
